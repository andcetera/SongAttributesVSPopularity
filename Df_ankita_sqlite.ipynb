```python
csv_data2 = "artist_genres.csv"
genre_df = pd.read_csv(csv_data2)
genre_df
```

```python
genre_df = genre_df.loc[genre_df['artist_genre'] != '0']
```

```python
most_pop_genres_df = pd.DataFrame(genre_df.groupby(genre_df['artist_genre']).count())

most_pop_genres_df = most_pop_genres_df.sort_values(['artist_names'], ascending=False)
top_fifty_genres_df = most_pop_genres_df.head(50)

top_fifty_genres_df
```

```python
aver_rank_streams = music_df.groupby(['artist_names','track_name'])['rank','streams'].mean().round(0)
aver_rank_streams_df = pd.DataFrame(aver_rank_streams)
#music_df.groupby(music_df['country'])['danceability', 'energy','loudness',
       #'speechiness', 'acousticness','valence', 'tempo', 'duration'].mean()
aver_rank_streams_df
```

```python
aver_rank_streams_df = aver_rank_streams_df.reset_index()
aver_rank_streams_df = aver_rank_streams_df[['artist_names','track_name','rank','streams']]
aver_rank_streams_df 
```

```python
max_weeks_on_chart = music_df.groupby(['artist_names','track_name'])['weeks_on_chart'].max()
max_weeks_on_chart_df = pd.DataFrame(max_weeks_on_chart).sort_values(['weeks_on_chart'], ascending=False)
max_weeks_on_chart_df = max_weeks_on_chart_df.reset_index()
```

```python
max_weeks_on_chart_df.columns
```

```python
coolness_df = pd.merge(max_weeks_on_chart_df,aver_rank_streams_df,how='left',left_on=['artist_names','track_name'],right_on=['artist_names','track_name'])
```

```python
coolness_df['WOC_ratio'] = coolness_df['weeks_on_chart'].div(290).round(3)
coolness_df.head(40)
```

```python
coolness_df['rank'].max()
```

```python
rank_pts = []
for row in coolness_df['rank']:
    rank_pts.append((100 - (row * .5))/100)
    
```

```python
coolness_df['rank_pts'] = rank_pts
```

```python
coolness_df.sort_values(['rank_pts'],ascending=False)
```

```python
coolness_df['streams_ratio'] = coolness_df['streams'].div(4026918).round(3)
coolness_df.sort_values(['streams_ratio'],ascending=False)
```

```python
coolness_df['Popularity_Score'] = (coolness_df['WOC_ratio'] + coolness_df['rank_pts'] + coolness_df['streams_ratio'])/3
```

```python
coolness_df = coolness_df.sort_values(['Popularity_Score'],ascending=False)
coolness_df.head(5)
```

```python
usa_df = usa_df.drop_duplicates(subset=['artist_names','track_name'],keep='first')
usa_df = usa_df[['artist_names','track_name','country']]
usa_df
```

```python
#import all dependencies for Sqlalchemy
from sqlalchemy import create_engine
import sqlite3

# Import modules to declare columns and column data types
from sqlalchemy import Column, Integer, String, DECIMAL, DateTime



# Import and establish Base for which classes will be constructed 
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()

engine = create_engine('sqlite://', echo=False)
```

```python
class chart_rank(Base):
    __tablename__ = "chart_rank_db"
    song_id = Column(Integer, primary_key=True)
    rank = Column(Integer)
    country = Column(String, primary_key=True)
    weeks_on_chart = Column(Integer)
    week = Column(DateTime, primary_key=True)
    artist_id = Column(Integer, foreign_key = True)
```

```python
# Create Database Connection
engine = create_engine('sqlite:///chart_rank_db.sqlite')
Base.metadata.create_all(engine)
```

```python
class song(Base):
    __tablename__ = "song_db"
    song_id = Column(Integer, primary_key=True)
    track_name = Column(String)
    streams = Column(Integer)
    artist_id = Column(Integer, foreign_key = True)
```

```python
class song_attributes(Base):
    __tablename__ = "attributes_db"
    song_id = Column(Integer, foreign_key=True)
    danceability = Column(DECIMAL)
    energy = Column(DECIMAL)
    loudness = Column(DECIMAL)
    speechiness = Column(DECIMAL)
    acousticness = Column(DECIMAL)
    valence = Column(DECIMAL)
    tempo =  Column(DECIMAL)
    duration = Column(DECIMAL)
```
