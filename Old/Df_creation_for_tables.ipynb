```python
#Dataframe creation by Crystal Butler
#import in cleaned song dataset created from "Explore_data.ipynb"
import pandas as pd
csv_data = "../Resources/dupl_removed_final3.csv"
```

```python
#create music df
music_df = pd.read_csv(csv_data)
```

```python
#view music df
music_df.head()
```

```python
# this will replace "United States" with "United States of America" and "Hong Kong" with "China"
music_df = music_df.replace(to_replace="United States",
           value="United States of America")
music_df = music_df.replace(to_replace="Hong Kong",
           value="China")
```

```python
#create groupby data for countries attributes df
countries_data = music_df.groupby(music_df['country'])['danceability', 'energy','loudness',
       'speechiness', 'acousticness','valence', 'tempo', 'duration'].mean()

```

```python
#create countries attrib df and reset index
countries_attrib_df = pd.DataFrame(countries_data).sort_values(['country']).round(2)
countries_attrib_df = countries_attrib_df.reset_index()
countries_attrib_df.head(5)
```

```python
#create countries list to be used to create another countries top 50 df
countries = []
countries = [column for column in countries_attrib_df['country']]
print(countries)
```

```python
#bring in artists_genre csv from dataset created from "Explore_data.ipynb"
csv_data2 = "../Resources/artist_genres.csv"
genre_df = pd.read_csv(csv_data2)
genre_df.head(5)
```

```python
#create genre dataframe from the csv
genre_df = genre_df.loc[genre_df['artist_genre'] != '0']
```

```python
#aggregate instances of genres across the dataframe 
most_pop_genres_df = pd.DataFrame(genre_df.groupby(genre_df['artist_genre']).count())
most_pop_genres_df = most_pop_genres_df.sort_values(['artist_names'], ascending=False)
#just keep the top 50 genres
top_fifty_genres_df = most_pop_genres_df.head(50)

top_fifty_genres_df.head(5)
```

```python
top_fifty_genres_df = top_fifty_genres_df.reset_index()
top_fifty_genres_df
```

```python
music_df.columns
```

```python
#find the average rank and number of streams of each song and artist combo
aver_rank_streams = music_df.groupby(['artist_names','track_name'])['rank','streams'].mean().round(0)
aver_rank_streams_df = pd.DataFrame(aver_rank_streams)

aver_rank_streams_df.head(5)
```

```python
#reset index
aver_rank_streams_df = aver_rank_streams_df.reset_index()
aver_rank_streams_df = aver_rank_streams_df[['artist_names','track_name','rank','streams']]
aver_rank_streams_df.head(5)
```

```python
#get max weeks on chart and convert to dataframe, reset index
max_weeks_on_chart = music_df.groupby(['artist_names','track_name'])['weeks_on_chart'].max()
max_weeks_on_chart_df = pd.DataFrame(max_weeks_on_chart).sort_values(['weeks_on_chart'], ascending=False)
max_weeks_on_chart_df = max_weeks_on_chart_df.reset_index()
```

```python
max_weeks_on_chart_df.columns
```

```python
#create coolness dataframe by merging max weeks and aver rank streams dataframes
coolness_df = pd.merge(max_weeks_on_chart_df,aver_rank_streams_df,how='left',left_on=['artist_names','track_name'],right_on=['artist_names','track_name'])
```

```python
#create a Weeks on Chart ratio to get a value to calculate popularity and add as a column to coolness df
coolness_df['WOC_ratio'] = coolness_df['weeks_on_chart'].div(290).round(3)
coolness_df.head(5)
```

```python
#find max rank
coolness_df['rank'].max()
```

```python
#iterate through rank values and create "rank pts" to get a value to calculate popularity
rank_pts = []
for row in coolness_df['rank']:
    rank_pts.append((100 - (row * .5))/100)
    
```

```python
#add rank pts to dataframe
coolness_df['rank_pts'] = rank_pts
```

```python
#cheaking rank pts values
coolness_df.sort_values(['rank_pts'],ascending=False).head(5)
```

```python
#create streams ratio and add to dataframe, so we can calculate popularity
coolness_df['streams_ratio'] = coolness_df['streams'].div(4026918).round(3)
coolness_df.sort_values(['streams_ratio'],ascending=False).head(5)
```

```python
#create a popularity score that averages the WOC ratio, rank pts, and streams ratio
coolness_df['Popularity_Score'] = (coolness_df['WOC_ratio'] + coolness_df['rank_pts'] + coolness_df['streams_ratio'])/3
```

```python
#sort Popularity score to see highest values.. helps find "most popular" song
coolness_df = coolness_df.sort_values(['Popularity_Score'],ascending=False)
coolness_df.head(5)
```

```python
usa_df = music_df.loc[music_df['country'] == 'United States of America']
usa_df.head(5)
```

```python
usa_df = usa_df.drop_duplicates(subset=['artist_names','track_name'],keep='first')
usa_df = usa_df[['artist_names','track_name','country']]
usa_df.columns = usa_df.columns.str.upper()
usa_df.head(5)
```

```python
usa_coolness_df = pd.merge(coolness_df,usa_df,how='inner',left_on=['artist_names','track_name'],right_on=['ARTIST_NAMES','TRACK_NAME'])
usa_coolness_df = usa_coolness_df.sort_values(['Popularity_Score'],ascending=False)
usa_coolness_df.head(5)
```

```python
x = usa_coolness_df['weeks_on_chart'].max() 

usa_coolness_df['weeks_on_chart'].div(x).head(5)
```

```python
# applying groupby() function to
# group the data on team value.
gk = music_df.groupby('country')
  
def create_df(country):
    cntry_df = gk.get_group(country)
    song_grpd = cntry_df.groupby(['artist_names','track_name']).aggregate({'streams':'mean','rank':'mean','weeks_on_chart':'max','danceability':'mean', 'energy':'mean','loudness':'mean',
        'speechiness':'mean', 'acousticness':'mean','valence':'mean', 'tempo':'mean', 'duration':'mean'}).round(3)
    song_grpd = song_grpd.reset_index()
    song_grpd['country'] = country
    song_grpd.sort_values(['weeks_on_chart'],ascending=False)
    x= song_grpd['weeks_on_chart'].max()
    song_grpd['WOC_ratio'] = song_grpd['weeks_on_chart'].div(x).round(3)
    song_grpd['rank'].max()
    rank_pts = []
    for row in song_grpd['rank']:
        rank_pts.append((100 - (row * .5))/100)
    song_grpd['rank_pts'] = rank_pts
    song_grpd.sort_values(['streams'],ascending=False)
    y = song_grpd['streams'].max()
    song_grpd['streams_ratio'] = song_grpd['streams'].div(y).round(3)
    song_grpd.sort_values(['streams_ratio'],ascending=False)
    song_grpd['Popularity_Score'] = (song_grpd['WOC_ratio'] + song_grpd['rank_pts'] + song_grpd['streams_ratio'])/3
    song_grpd.columns
    best_songs = song_grpd[['artist_names', 'track_name', 'Popularity_Score', 'WOC_ratio', 'rank_pts',
        'streams_ratio','streams', 'rank', 'weeks_on_chart',
        'danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
        'valence', 'tempo', 'duration','country']]
    best_songs = best_songs.sort_values(['Popularity_Score'],ascending=False)
    top_fifty = best_songs.head(50)
    
    return top_fifty
```

```python
first = True
for c in countries:
    country_df = create_df(c)
    if first:
        all_countries_top_fifty_df = country_df
        first = False
    all_countries_top_fifty_df = pd.merge(all_countries_top_fifty_df,country_df, how='outer')
```

```python
#Christin adding primary keys to sqlite tables
#add primary key to country_attributes (1)

#connect to the database
conn = sqlite3.connect('songs_complete_db.sqlite')
c = conn.cursor()

c.executescript('''
    PRAGMA foreign_keys=off;

    BEGIN TRANSACTION;
    ALTER TABLE country_attributes RENAME TO old_table;

    /*create a new table with the same column names and types while
    defining a primary key for the desired column*/
    CREATE TABLE country_attributes (
      country TEXT PRIMARY KEY NOT NULL,
      danceability REAL,
      energy REAL,
      loudness REAL,
      speechiness REAL,
      acousticness REAL,
      valence REAL,
      tempo REAL,
      duration REAL);

    INSERT INTO country_attributes SELECT country, danceability, energy, loudness, speechiness, acousticness, valence, tempo, duration FROM old_table;

    DROP TABLE old_table;
    COMMIT TRANSACTION;

    PRAGMA foreign_keys=on;''')

#close out the connection
c.close()
conn.close()

time.sleep(5)
```

```python
#add primary key to coolness (2)

#connect to the database
conn = sqlite3.connect('songs_complete_db.sqlite')
c = conn.cursor()

c.executescript('''
    PRAGMA foreign_keys=off;

    BEGIN TRANSACTION;
    ALTER TABLE coolness RENAME TO old_table;

    /*create a new table with the same column names and types while
    defining a primary key for the desired column*/
    CREATE TABLE coolness (
    artist_names TEXT,
    track_name TEXT,
    weeks_on_chart INTEGER,
    rank REAL,
    streams REAL,
    WOC_ratio REAL,
    rank_pts REAL,
    streams_ratio REAL,
    Popularity_Score REAL,
    PRIMARY KEY (artist_names, track_name)
    );

    INSERT INTO coolness SELECT artist_names, track_name, weeks_on_chart, rank, streams, WOC_ratio, rank_pts, streams_ratio, Popularity_Score FROM old_table;

    DROP TABLE old_table;
    COMMIT TRANSACTION;

    PRAGMA foreign_keys=on;''')

#close out the connection
c.close()
conn.close()

time.sleep(5)
```

```python
#add primary key to fiftygenre (3)

#connect to the database
conn = sqlite3.connect('songs_complete_db.sqlite')
c = conn.cursor()

c.executescript('''
    PRAGMA foreign_keys=off;

    BEGIN TRANSACTION;
    ALTER TABLE fiftygenre RENAME TO old_table;

    /*create a new table with the same column names and types while
    defining a primary key for the desired column*/
    
    CREATE TABLE fiftygenre (
        artist_genre TEXT PRIMARY KEY NOT NULL,
        artist_names INTEGER
    );

    INSERT INTO fiftygenre SELECT artist_genre, ARTIST_NAMES FROM old_table;

    ALTER TABLE fiftygenre RENAME COLUMN ARTIST_NAMES TO ARTIST_COUNT;

    DROP TABLE old_table;

    COMMIT TRANSACTION;

    PRAGMA foreign_keys=on;''')

#close out the connection
c.close()
conn.close()

time.sleep(5)
```

```python
#add primary key to country and rename to top_50_by_country (4)

#connect to the database
conn = sqlite3.connect('songs_complete_db.sqlite')
c = conn.cursor()

c.executescript('''
    PRAGMA foreign_keys=off;

    BEGIN TRANSACTION;
    ALTER TABLE country RENAME TO old_table;

    /*create a new table with the same column names and types while
    defining a primary key for the desired column*/
    DROP TABLE IF EXISTS top_50_by_country;
    CREATE TABLE top_50_by_country (
        artist_names TEXT,
        track_name TEXT,
        Popularity_Score REAL,
        WOC_ratio REAL,
        rank_pts REAL,
        streams_ratio REAL,
        streams REAL,
        rank REAL,
        weeks_on_chart INTEGER,
        danceability REAL,
        energy REAL,
        loudness REAL,
        speechiness REAL,
        acousticness REAL,
        valence REAL,
        tempo REAL,
        duration REAL,
        country TEXT,
        PRIMARY KEY (artist_names, track_name, country)
    );

    INSERT INTO top_50_by_country SELECT artist_names, track_name, Popularity_Score, WOC_ratio, rank_pts, streams_ratio, streams, rank, weeks_on_chart, danceability, energy, loudness, speechiness, acousticness, valence, tempo, duration, country FROM old_table;

    DROP TABLE old_table;
    COMMIT TRANSACTION;

    PRAGMA foreign_keys=on;''')

#close out the connection
c.close()
conn.close()

time.sleep(5)
```
