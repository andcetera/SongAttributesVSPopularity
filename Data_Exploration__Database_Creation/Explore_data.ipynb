```python
#Data Exploration by Crystal Butler

#Pull in song data from final.csv file
import pandas as pd
csv_data = "final.csv"
```

```python
#read in csv data to dataframe
music_df = pd.read_csv(csv_data)
```

```python
#view dataframe
music_df.head(5)
```

```python
#seeing what column names are included in the dataframe
music_df.columns
```

```python
#take away unneeded columns and create new reduced df
music_reduced = music_df[['artist_names', 'track_name','rank', 'artists_num', 'artist_genre', 'release_date','weeks_on_chart',
       'streams', 'week', 'danceability', 'energy','loudness',
       'speechiness', 'acousticness', 'instrumentalness',
       'valence', 'tempo', 'duration', 'country']]
```

```python
#view new dataframe
music_reduced.head()
```

```python
#check out datatypes of each column
music_reduced.dtypes
```

```python
#delete rows that have strings in them
music_removed = music_reduced[music_reduced['rank'].str.contains("rank") == False]
```

```python
#show unique values for the rank and remove unnecessary spaces
music_removed['rank'].str.strip()
music_removed['rank'].unique()

```

```python

# creating a dictionary 
# with column name and data type

data_types_dict = {'artist_names': str, 'track_name': str, 'rank': int,'artists_num': float, 'artist_genre': str,'weeks_on_chart': float,
       'streams': float, 'danceability': float, 'energy': float,'loudness': float,
       'speechiness': float, 'acousticness': float, 'instrumentalness': float,
       'valence': float, 'tempo': float, 'duration': float, 'country': str}
  
# we will change the data type 
# of id column to str by giving
# the dict to the astype method
# change two date columns from string to datatime dtype
music_removed = music_removed.astype(data_types_dict)
music_removed["release_date"] = pd.to_datetime(music_removed["release_date"])
music_removed["week"] = pd.to_datetime(music_removed["week"])  
# checking the data types
# using df.dtypes method
music_removed.dtypes
```

```python
#view changes
music_removed.head()
```

```python
#seeing if we have rows with null data

music_removed.count()
```

```python
#find all unique values for "week" column
unique_dates = music_removed['week'].unique()
len(unique_dates)
```

```python
#put these dates into a list and sort, so we can see the first and last date
unique_dates = [unique for unique in unique_dates]
len(unique_dates)
unique_dates.sort()
print(unique_dates[0])
print(unique_dates[-1])
```

```python
#grouping songs and seeing which countries that they have streams for
artist_group = music_removed.groupby(music_removed['artist_names'])
artist_grouped = music_removed.groupby(["artist_names", "track_name","country"])["streams"].sum()
artist_grouped_df = pd.DataFrame(artist_grouped).sort_values(['track_name','streams'])
artist_grouped_df.head(5)
```

```python
#seeing all the unique countries in the database
music_removed['country'].unique()
```

```python
#removing "global" from dataframe because we only want unique 'country' values
df_w_countries = music_removed.loc[music_removed['country'] != 'Global']
df_w_countries['country'].unique()
df_w_countries.head()
```

```python
#looking at songs grouped with the countries that it is streamed in without 'global'
countries_grouped =df_w_countries.groupby(["artist_names", "track_name","country"])["streams"].sum()
df_countries_grouped = pd.DataFrame(countries_grouped).sort_values(['track_name','streams'])
df_countries_grouped.head(5)
```

```python
#only pulling songs where United States is the country
usa_df = music_removed.loc[music_removed['country'] == 'United States']
usa_df.head()
```

```python
#seeing how many unique dates are in the usa_df
dates_unique = usa_df['week'].unique()
len(dates_unique)
```

```python
#dropping null rows
df_w_countries = df_w_countries.dropna(how='any')
df_w_countries.head()
```

```python
#creating csv file
df_w_countries.to_csv('musicdf1.csv',index=False)
```

```python
#checking how many columns are in the countries df
len(df_w_countries.columns)
```

```python
#creating artist genre df
artist_genre_df = df_w_countries[['artist_names','artist_genre']]
artist_genre_df.head()
```

```python
#sending artist genre df to csv
artist_genre_df.to_csv('artist_genres.csv',index=False)
```

```python
#getting rid of columns we don't need in this df
dropped_genre_df = df_w_countries.drop(df_w_countries[['artist_genre','release_date','artists_num','instrumentalness']], axis=1)

dropped_genre_df.head()
```

```python
#dropping duplicate row values
dup_removed_df = dropped_genre_df.drop_duplicates(subset=['artist_names','track_name','country','week'],keep='first').sort_values(['track_name','country','week'])
dup_removed_df
```

```python
#creating csv for dup removed df
dup_removed_df.to_csv('dupl_removed_final.csv',index=False)
```
