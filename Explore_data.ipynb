```python
music_reduced = music_df[['artist_names', 'track_name','rank', 'artists_num', 'artist_genre', 'release_date','weeks_on_chart',
       'streams', 'week', 'danceability', 'energy', 'key', 'loudness',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'duration', 'country']]
```

```python
#music_removed = music_reduced[~music_reduced.rank.str.contains("rank")]
music_removed = music_reduced[music_reduced['rank'].str.contains("rank") == False]
```

```python
music_removed['rank'].str.strip()
music_removed['rank'].unique()

#points = df.points.unique()
```

```python

# creating a dictionary 
# with column name and data type
#data_types_dict = {'artist_names': str, 'track_name': str, 'rank': int,'artists_num': float, 'artist_genre': str, 'peak_rank': int, 'previous_rank': int, 'weeks_on_chart': int,
       #'streams': int, 'danceability': float, 'energy': float, 'key': int, 'mode': int, 'loudness': float,
       #'speechiness': float, 'acousticness': float, 'instrumentalness': float, 'liveness': float,
       #'valence': float, 'tempo': float, 'duration': int, 'country': str, 'region': str}
data_types_dict = {'artist_names': str, 'track_name': str, 'rank': int,'artists_num': float, 'artist_genre': str,'weeks_on_chart': float,
       'streams': float, 'danceability': float, 'energy': float, 'key': float,'loudness': float,
       'speechiness': float, 'acousticness': float, 'instrumentalness': float, 'liveness': float,
       'valence': float, 'tempo': float, 'duration': float, 'country': str}
  
# we will change the data type 
# of id column to str by giving
# the dict to the astype method
music_removed = music_removed.astype(data_types_dict)
  
# checking the data types
# using df.dtypes method
music_removed.dtypes
```

```python
music_removed["release_date"] = pd.to_datetime(music_removed["release_date"])
music_removed["week"] = pd.to_datetime(music_removed["week"])
```
